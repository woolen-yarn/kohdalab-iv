# KohdaLab IV Measurement

DC IV/VI sweep helper notebook.

```python
from kohdalab_iv.api import Experiment, iv_plan_from_config, load_config
from kohdalab_iv.api.config import resolve_config_path
from kohdalab_iv.api.notebook import format_point, make_iv_live_update

config_path = resolve_config_path().path
if config_path is None:
    raise FileNotFoundError('No configuration file could be resolved.')
config = load_config(config_path)
plan = iv_plan_from_config(config)
experiment = Experiment(config, auto_connect=False)
print(plan.summary)
print(plan.source_ref, plan.source_model)
print(plan.measure_ref, plan.measure_model)
```

```python
from kohdalab_iv.interfaces.common import list_visa_resources

list_visa_resources()
```

```python
live_update = make_iv_live_update(x_key='measured_V', y_key='measured_A')

def on_point(point):
    print(format_point(point))
    live_update(point)

# This cell talks to the real instruments and writes CSV continuously.
rows = experiment.run_iv(plan=plan, on_point=on_point)
len(rows)
```
